I chose a dataset on the happiness of people worldwide, this study is based on the annual report from 2017. This report ranks 155 countries by their happiness levels, this was released at the United Nations at an event celebrating International Day of Happiness on March 20th. The columns: Country or region, GDP per capita, Social support, Healthy life expectancy, Freedom to make life choices, Generosity, Perceptions of corruption describe the extent to which these factors contribute in evaluating the happiness in each country. A happiness score in general is a subjective evaluation, in a capitalistic country people would be happier with the attainment of wealth and objects, while in another happiness may be driven by family values and another associating happiness with love. Freedom also is an incredibly subjective concept, to be free in the United States is very different from being free in China, for example: some Americans may associate their freedom with gun ownership while those in China associate it with protesting rights or privacy rights. Perceptions of corruption is another socially constructed topic because corruption has different interpretations in different countries, for example in the United States lobbying is popular, but in other countries corruption may look like private transactions or unauthorized militant actions. What has a larger influence on Happiness score, GDP per capita or perceptions of corruption? This will be answered by measuring the correlation between these variables and the happiness score and comparing the results.

```python
#EXTRA CREDIT first attempt (ignore linear regression models, look to spearman regression model)
plt.scatter(df['GDP per capita'], df['Perceptions of corruption'], alpha=0.5)
plt.title('Linear Regression Plot')
plt.xlabel('GDP per capita')
plt.ylabel('Perceptions of corruption')
plt.show()
#no strong correlation emerges, possibly a weak positive correlation, 
#but it is notable that those with high GDP tend to also have high perceptions of corruption
```

```python
plt.scatter(df['Perceptions of corruption'], df['GDP per capita'], alpha=0.5)
plt.title('Linear Regression Plot')
plt.xlabel('Perceptions of corruption')
plt.ylabel('GDP per capita')
plt.show()
#once again no strong correlation emerges possibly a weak positive correlation, 
#and those with high GDP tend to also have high perceptions of corruption
```

```python
import seaborn as sns
from scipy.stats import spearmanr
# Calculate Spearman correlation coefficient and p-value
corr, pval = spearmanr(df['GDP per capita'], df['Perceptions of corruption'])

# Print results
print('Spearman correlation coefficient:', corr)
print('P-value:', pval)
sns.regplot(x='GDP per capita', y='Perceptions of corruption', data=df, color='b', scatter_kws={'s': 50})
sns.lineplot(x='GDP per capita', y='Perceptions of corruption', data=df, color='r')

# Add correlation coefficient and p-value to plot
plt.text(df['GDP per capita'].min(), df['Perceptions of corruption'].max(), f'Spearman correlation = {corr:.2f}\np-value = {pval:.2f}', fontsize=12, verticalalignment='top')

# Show plot
plt.show()
```
